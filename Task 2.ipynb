```python
import pandas as pd

df = pd.read_csv("../../Dataset.csv")

print(df.head())
```

```python
common_price_range = df['Price range'].mode()[0]
print("Most Common Price Range:", common_price_range)

print(df['Price range'].value_counts())
```

```python
avg_rating_by_price = df.groupby("Price range")["Aggregate rating"].mean()
print("\nAverage Rating by Price Range:\n", avg_rating_by_price)
```

```python
avg_rating_color = df.groupby(["Price range", "Rating color"])["Aggregate rating"].mean()

# Reset index for readability
avg_rating_color = avg_rating_color.reset_index()

print("\nAverage Rating by Price Range and Rating Color:\n", avg_rating_color)

# Find the top rating color overall (highest average)
top_color = avg_rating_color.loc[avg_rating_color["Aggregate rating"].idxmax()]
print("\nColor representing highest average rating:")
print(top_color)
```

```python
import matplotlib.pyplot as plt

# Price range distribution
df['Price range'].value_counts().plot(kind='bar')
plt.title("Most Common Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()

# Average rating by price range
avg_rating_by_price.plot(kind='bar')
plt.title("Average Rating by Price Range")
plt.ylabel("Average Rating")
plt.show()
```
